```python
processed_text=""
with open("output_text.txt",'r',encoding='UTF-8') as input_text:
    processed_text=input_text.read()
print(processed_text)
```

```python
text_list=processed_text.split('\n\n')
text_list
```

```python
import re

sep_list=[]
for i,ele in enumerate(text_list):
    crr=[]
    for lin in re.sub(r'^[0-9]+\.|[abcd]\)',"",ele).split('\n'):
        crr.append(lin.strip())
    if not len(crr)==5:
        print(f"{i+1} -> {crr}")
    sep_list.append(crr)
    
```

```python
sep_list.pop()
sep_list
```

```python
question_list=[]
for i,ele in enumerate(sep_list):
    ques_dict=dict()
    ques_dict["question"]=ele[0]
    ques_dict['options']=ele[1:]
    question_list.append(ques_dict)
    
```

```python
import json

with open("question_set_1.json","w") as f:
    json.dump(question_list,f,indent=4)
```
