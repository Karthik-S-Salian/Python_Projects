```python
import pandas as pd 
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv("../resources/Excel_sheets/sonar_dataset.csv",header=None)
df.sample(4)
```

```python
df[60].replace({"R":0,"M":1},inplace=True)
```

```python
X=df.drop(60,axis=1)
y=df[60]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=5)
X_train.shape[0]
```

```python
model=keras.Sequential([
keras.layers.Dense(60,input_dim=X_train.shape[1],activation='relu'),
keras.layers.Dropout(0.5),
keras.layers.Dense(30,activation='relu'),
keras.layers.Dropout(0.5),
keras.layers.Dense(15,activation='relu'),
keras.layers.Dropout(0.25),
keras.layers.Dense(1,activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train,y_train,epochs=50,batch_size=8)
```

```python
print(classification_report(y_test,np.round(model.predict(X_test))))
```
