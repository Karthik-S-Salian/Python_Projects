```python
def binary_search(search_word,search_list):
    begin=0
    end=len(search_list)
    found=False
    
    while(begin<=end):
        mid=(begin+end)//2
        for i in range(5):
            if search_list[mid][i]==search_word[i]:
                found=True
                continue
            if search_list[mid][i]>search_word[i]:
                found=False
                end=mid-1
                break
            if search_list[mid][i]<search_word[i]:
                found=False
                begin=mid+1
                break
        if found:
            return mid
    return None
```

```python
from random import randint
list_5_letters.sort()
    
```

```python
l=len(list_5_letters)-1
for i in range(0,100):
    real_loc=randint(0,l)
    search_word=list_5_letters[real_loc]
    status=binary_search(search_word,list_5_letters)
    if status is None:
        print(f"{real_loc = }  {search_word = }  {status} \n")
```
