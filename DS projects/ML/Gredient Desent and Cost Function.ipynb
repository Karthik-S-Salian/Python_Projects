cost function = sum((y-ypredicted)^2)/n =  sum((y-m*x+b)^2)/n

```python
def gredient_descent(x,y):
    m_min=m_crr=2.571795
    b_min=b_crr=0
    iterations=1000
    n=len(x)
    learning_rate=0.001
    y_predicted=m_crr*x+b_crr
    md_min=abs((2/n)*sum(x*(y-y_predicted)))
    bd_min=abs((2/n)*sum(y-y_predicted))
    
    for i in range(iterations):
        y_predicted=m_crr*x+b_crr
        md=-(2/n)*sum(x*(y-y_predicted))
        bd=-(2/n)*sum(y-y_predicted)
        #m_crr=m_crr-learning_rate*md
        if(abs(md)<md_min):
            m_min=m_crr
            md_min=abs(md)
        if(abs(bd)<bd_min):
            b_min=b_crr 
            bd_min=abs(bd)
        b_crr=b_crr-learning_rate*bd
        print(f'm = {m_crr: 3f} , b_crr ={b_crr: 3f} , md = {md: 3f} , bd = {bd} , iteration = {i} ')
    print(f'{md_min = } , {bd_min = }')
    print(f'{m_min = } , {b_min = }')
```
